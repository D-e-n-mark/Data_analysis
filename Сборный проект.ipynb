**–§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö**

```python
def vis_nan(df):
    # –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –ø—Ä–æ–ø—É—Å–∫–∞—Ö –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö –ø–æ –∫–∞–∂–¥–æ–º—É —Å—Ç–æ–ª–±—Ü—É
    for col in df.columns:
        pct_missing = np.mean(df[col].isnull())
        print('{} - {}%'.format(col, round(pct_missing*100)))
    
    # –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–∞ —Ö–∏—Ç–º—ç–ø–µ
    colours = ['black', '#b4fa4b'] 
    fig, ax = plt.subplots(figsize=(12,12))
    sns.heatmap(df.isnull(), yticklabels=False, cmap=sns.color_palette(colours))
    
    # –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –∑–Ω–∞—á–µ–Ω–∏–π –ª–µ–≥–µ–Ω–¥—ã 
    colorbar = ax.collections[0].colorbar
    colorbar.set_ticks([1/4,3/4])
    colorbar.set_ticklabels(['–ï—Å—Ç—å –¥–∞–Ω–Ω—ã–µ','–ù–µ—Ç –¥–∞–Ω–Ω—ã–µ—Ö'])
```

**–§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–≤–æ–¥–∞ –≤—Å–µ–π –æ–±—â–µ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏**

```python
def gen_info(df):
    return display(df.sample(10)), \
    display(df.describe()), \
    display(df.info()), \
    display(df.shape), \
    display(vis_nan(df))
```

**–§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –ø–æ—Ç–µ—Ä—å –≤ –¥–∞–Ω–Ω—ã—Ö –≤ —Å—Ä–∞–≤–Ω–µ–Ω–∏–∏ —Å –∏—Å—Ö–æ–¥–Ω—ã–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–æ–º**

```python
def data_loss():
    percentage = ((games.shape[0] / initial_games.shape[0]) * 100)
    return round(percentage, 3)
```

**–§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–º —Å—Ç–æ–ª–±—Ü–µ**

```python
def nan_count(col):
    return games[col].isna().sum()
```
